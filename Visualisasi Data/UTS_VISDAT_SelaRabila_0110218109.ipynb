### 5. Lengkapilah kode dibawah ini

```python
# import library
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset iris and save to data_iris dataframe
data_iris =  sns.load_dataset("iris")

# SOAL 5.a
# Tuliskan kode program untuk informasi mengenai data pada dataframe data_iris
data_iris.info()

# SOAL 5.b
# Tuliskan kode program untuk melihat 5 data pertama pada dataframe data_iris
data_iris.head(5)

# SOAL 5.c
# Tuliskan kode program untuk menampilkan diagram batang jumlah masing-masing kelas(species)
data_iris['species'].value_counts()
data_iris['species'].value_counts().plot.hist(data_iris.hist(edgecolor='black', linewidth=1.2))
plt.show()

```

### 6. Kode dibawah ini merupakan kode untuk menampilkan scatter plot. Apa informasi yang bisa diambil dari grafik tersebut?

**jawaban:**
