```python
#import library
import matplotlib
from matplotlib import transforms, pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# Definisikan warna-warna yang akan digunakan
GRAY1, GRAY2, GRAY3 = '#231F20', '#414040', '#555655'
GRAY4, GRAY5, GRAY6 = '#646369', '#76787B', '#828282'
GRAY7, GRAY8, GRAY9 = '#929497', '#A6A6A5', '#BFBEBE'
BLUE1, BLUE2, BLUE3, BLUE4 = '#174A7E', '#4A81BF', '#94B2D7', '#94AFC5'
RED1, RED2 = '#C3514E', '#E6BAB7'
GREEN1, GREEN2 = '#0C8040', '#9ABB59'
ORANGE1 = '#F79747'
```

```python
# konfigurasi font yang akan digunakan menjadi Arial
plt.rcParams['font.family'] = 'Arial'
# konfigurasi teks tebal dan miring
matplotlib.rcParams['mathtext.fontset'] = 'custom'
matplotlib.rcParams['mathtext.bf'] = 'Arial:bold'
matplotlib.rcParams['mathtext.it'] = 'Arial:italic'
```

```python
# Siapkan data yang akan digunakan
# number of tickets per month
X = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
     'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Y = [160, 180, 250, 140, 175, 150, 125, 156, 126, 104, 124, 140] # processed
Z = [160, 180, 255, 140, 175, 160, 135, 202, 160, 139, 149, 177] # received
```

```python
# Buat gambar baru
fig, ax1 = plt.subplots(figsize=(8.2, 6.09), # Lebar, dan tinggi dalam satuan inchi
                        dpi=110)             # resolusi gambar
# Tuning posisi gambar
fig.subplots_adjust(left=0.104, right=0.768, top=0.751, bottom=0.187)
```

```python
# Gambar garis yang menunjukkan "received tickets"
ax1.plot(range(12), Z, linewidth=4, color=GRAY6)
ax1.scatter(range(12)[-5:], Z[-5:], s=90, color=GRAY6, clip_on=False)

#Tampilkan hasil
fig
```

```python
# Tampilkan garis yang menunjukkan "processed tickets"
ax1.plot(range(12), Y, linewidth=4, color=BLUE1)
ax1.scatter(range(12)[-5:], Y[-5:], s=90, color=BLUE1, clip_on=False)

#Tampilkan hasil
fig
```

```python
# Beri anotasi (nilai angka) pada garis "received"
for i, v in enumerate(Z[-5:]):
    ax1.annotate(str(v),
                 (i+7, v),                    # (x,y) Titik pada garis
                 xytext=(-11, 14),            # (x,y) yang akan diberi text
                 textcoords='offset points',  # offset (in points)
                 color=GRAY6,
                 fontsize=15)

#Tampilkan hasil
fig
```

```python
# Beri nilai pada garis "proccessed"
for i, v in enumerate(Y[-5:]):
    ax1.annotate(str(v),
                 (i+7, v),                   # (x,y) titik pada garis
                 xytext=(-11, -24),          # (x,y) posisi text
                 textcoords='offset points', # offset (in points)
                 color=BLUE1,
                 fontsize=15)

#Tampilkan hasil
fig
```

```python
# set batasan (limit) data untuk sumbu y dan x
ax1.set_xlim([0, 11])
ax1.set_ylim([0, 300])

#Tampilkan hasil
fig
```

```python
# pengaturan tulisan/titik pada sumbu
plt.setp(ax1,
         xticks=list(range(12)),
         xticklabels=X,
         yticks=list(range(0, 350, 50)))

#Tampilkan hasil
fig
```

```python
# Ubah tampilan label pada sumbu x, y
for i in ax1.get_yticklabels() + ax1.get_xticklabels():
    i.set_fontsize(14)
    i.set_color(GRAY4)

#Tampilkan hasil
fig
```

```python
# Geser tulisan pada sumbu y agak ke kiri
for item in ax1.get_yticklabels():
    offset = transforms.ScaledTranslation(-0.07, 0, fig.dpi_scale_trans)
    item.set_transform(item.get_transform() + offset)

#Tampilkan hasil
fig
```

```python
# ubah penampilan dari ticks, tick labels, dan gridlines
ax1.tick_params(bottom='off')

#Tampilkan hasil
fig
```

```python
# Ubah warna
ax1.tick_params(color=GRAY4)
ax1.spines['bottom'].set_color(GRAY4)
ax1.spines['left'].set_color(GRAY4)

#Tampilkan hasil
fig
```

```python
# hapus chart border
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)

#Tampilkan hasil
fig
```

# Tampilan text & garis

```python
# Buat Gambar
fig, ax1 = plt.subplots(figsize=(8.2, 6.09), # Lebar, tinggi dalam inches
                        dpi=110)             # resolusi gambar

# geser gambar
fig.subplots_adjust(left=0.104, right=0.768, top=0.751, bottom=0.187)

# gambar garis "received tickets"
ax1.plot(range(12), Z, linewidth=4, color=GRAY6)
ax1.scatter(range(12)[-5:], Z[-5:], s=90, color=GRAY6, clip_on=False)

# Gambar garis "processed tickets"
ax1.plot(range(12), Y, linewidth=4, color=BLUE1)
ax1.scatter(range(12)[-5:], Y[-5:], s=90, color=BLUE1, clip_on=False)

# beri tulisan pada garis "received"
for i, v in enumerate(Z[-5:]):
    ax1.annotate(str(v),
                 (i+7, v),                    # (x,y) titik yang akan ditulis
                 xytext=(-11, 14),            # (x,y) posisi text
                 textcoords='offset points',  # offset (in points)
                 color=GRAY6,
                 fontsize=15)

# beri tulisan pada garis "proccessed"
for i, v in enumerate(Y[-5:]):
    ax1.annotate(str(v),
                 (i+7, v),                   # (x,y) titik yang akan ditulis
                 xytext=(-11, -24),          # (x,y) posisi text
                 textcoords='offset points', # offset (in points)
                 color=BLUE1,
                 fontsize=15)


# set batasan data pada sumbu y dan x
ax1.set_xlim([0, 11])
ax1.set_ylim([0, 300])

# set tampilan sumbu
plt.setp(ax1,
         xticks=list(range(12)),
         xticklabels=X,
         yticks=list(range(0, 350, 50)))

# atur tampilan sumbu x, y
for i in ax1.get_yticklabels() + ax1.get_xticklabels():
    i.set_fontsize(14)
    i.set_color(GRAY4)

# gaset tulisan sumbu y sedikit ke kiri
for item in ax1.get_yticklabels():
    offset = transforms.ScaledTranslation(-0.07, 0, fig.dpi_scale_trans)
    item.set_transform(item.get_transform() + offset)

# ubah tampilan sumbu
ax1.tick_params(bottom='off')

# ganti warna
ax1.tick_params(color=GRAY4)
ax1.spines['bottom'].set_color(GRAY4)
ax1.spines['left'].set_color(GRAY4)

# hilangkan border pada grafik
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)

# tambahkan garis referensi vertikal
plt.axvline(x=4, ymax=0.8, color='#636369', linewidth=1)

# Tulis label pada tiap garis
ax1.text(11.5, 172, 'Received', fontsize='21', color=GRAY6)
ax1.text(11.5, 135, 'Processed', fontsize='21', color=BLUE1)


# Beri judul pada tiap sumbu
ax1.text(-1.6, 325, 'Ticket volume over time', fontsize=19, color=GRAY4)
ax1.text(-1.55, 150, 'Number of tickets', fontsize=14, rotation='vertical',
         color=GRAY4)
ax1.text(-0.34, -43, '2014', fontsize=14, color=GRAY4)

# Beri judul grafik
ax1.text(-1.6, 397, '$\\bf{Please\ approve\ the\ hire\ of\ 2\ FTEs}$',
         fontsize=25, color=GRAY4)
ax1.text(-1.6, 368, 'to backfill those who quit in the past year',
         fontsize=16, color=GRAY7)

# Tambahkan informasi dalam bentuk tulisan
ax1.text(4, 252,
         '$\\bf{2\ employees\ quit\ in\ May}$. We nearly kept up with'
         ' incoming\n'
         'volume in the following two months, but fell behind with the\n'
         'increase in Aug and haven''t been able to catch up since.',
         fontsize=14, color=GRAY5, linespacing=1.4)

# tambahkan footnote pada grafik (sumber data)
ax1.text(-1.6, -83,
         'Data source: XYZ Dashboard, as of 12/31/2014 | A detailed analysis'
         ' on tickets processed per person\n'
         'and time to resolve issues was undertaken to inform this request'
         ' and can be provided if needed.',
         fontsize=12.5, color=GRAY3, linespacing=1.3)
```
